```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#sns.set() → Global Theme Setup

penguins = sns.load_dataset('penguins')

sns.set()  # applies default seaborn style

df = sns.load_dataset("penguins")
sns.scatterplot(x="flipper_length_mm", y="body_mass_g", data=penguins)

plt.show()


# sns.set() applies a default clean theme to all plots
# improves grid, colors, and readability automatically
```

```python
#sns.set_style() → Change Background Style


# set_style controls background look
# "whitegrid" → white background + grid (most used)
# "darkgrid" → dark background + grid
# "white" → clean without grid
# "ticks" → minimal style with axis ticks


sns.set_style("whitegrid")  # try: darkgrid, white, ticks

sns.boxplot(x="species", y="body_mass_g", data=penguins)
plt.show()
```

```python
#sns.set_palette() → Change Colors 
# set_palette changes color scheme of plots
# "Set2" → soft colors (good for presentations)
# "deep" → default strong colors
# "pastel" → light colors

sns.set_palette("Set2")  # try: deep, muted, pastel

sns.barplot(x="species", y="body_mass_g", data=penguins)
plt.show()
```

```python
#sns.despine() → Remove Borders

# despine removes extra borders (top/right)
# makes plot cleaner and more professional

sns.set_style("white")

sns.histplot(x="body_mass_g", data=penguins)

sns.despine()  # removes top and right borders
#left = True removes left sns.despine(left =True) 

plt.show()
```

```python
#combinely  for  practice 
sns.set(style="whitegrid")
sns.set_palette("Set2")

sns.boxplot(x="species", y="body_mass_g", data=penguins)

sns.despine()

plt.show()
```
