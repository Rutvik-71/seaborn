```python
import pandas as pd
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# to check the names of the available datasets 

print(sns.get_dataset_names())
```

```python
penguins = sns.load_dataset('penguins')
```

```python
penguins.head()
```

```python
# 1) line chart 
#Lineplot is used to show the relationship between two continuous variables. 
#It is commonly used for time-series or ordered data.
sns.load_dataset('penguins')
sns.lineplot(x ="body_mass_g",y ="bill_length_mm",hue = "sex",data =penguins)
plt.show()
```

```python
# 2) scatterplot 
#Scatterplot is used to show the relationship between two numerical variables. 
#It helps in identifying patterns or correlations.

sns.scatterplot(x = 'flipper_length_mm', y = 'body_mass_g', data = penguins,hue = 'species')
```

```python
# 3) Barplot
#Barplot is used to compare numerical values across categories. 
#It shows the average value with confidence intervals.

sns.barplot(x='body_mass_g',y = 'species' ,data = penguins)
plt.show()
```

```python
#4) BoxPlot
#Boxplot is used to show the distribution of data based on quartiles. 
#It helps identify spread and outliers.
#A boxplot needs:
#one categorical column (x-axis)
#one numerical column (y-axis)

sns.boxplot(x= 'species' ,y ='body_mass_g' ,data = penguins)
```

```python
#5) Histplot
#Histplot is used to visualize the distribution of a single numerical variable.
#histplot uses only numerical data (no categorical on x like boxplot)
sns.histplot(x = 'body_mass_g',data = penguins ,kde = True)#kde=True adds density curve
```

```python
# 6) Heatmaps
#Heatmap is used to visualize matrix-like data using colors. 
#It is commonly used for correlation matrices.


# load dataset
df = sns.load_dataset("penguins")

# create correlation matrix (only numbers)
corr = df.corr(numeric_only=True)

# draw heatmap
sns.heatmap(corr, annot=True)

# show plot
plt.show()
```

```python
# cmap (color map) → controls colors in heatmap

# Types of colormaps:

# 1. Diverging (for data with negative & positive values)
#    examples: "coolwarm", "RdBu", "seismic"
#    use: correlation matrix (best choice)

# 2. Sequential (for data from low → high only)
#    examples: "viridis", "plasma", "Blues"
#    use: distributions, intensity, counts

# 3. Qualitative (for categories)
#    examples: "Set1", "tab10", "Pastel1"
#    use: categorical data (rare in heatmaps)

# Tip: use "coolwarm" or "RdBu" for correlation heatmaps
```

```python
# 7) PairPlot  
# Pairplot is used to plot pairwise relationships between multiple numerical variables.
iris = sns.load_dataset("penguins")
sns.pairplot(penguins, hue="species")
plt.show()
```

```python
# 8) Count Plot 
#It shows the distribution of a single categorical variable or the relationship between two categorical variables by creating a bar plot.


sns.countplot(x = 'sex' ,data = penguins)
plt.show()
```

```python
# Violin plot 
#A violin plot combines a box plot with a density plot, showing both summary stats and distribution shape
#Applications
'''Comparing distributions more deeply than boxplot.
Helpful for detecting multimodal distributions.'''

sns.violinplot(x="species", y="body_mass_g", data=penguins, hue="sex", split=True)
plt.title("violin plot")
plt.show()
```

```python
# KDE Plot 
#KDE Plot described as Kernel Density Estimate is used for visualizing the Probability Density of a continuous variable.

sns.kdeplot(x="body_mass_g", data=penguins, hue="species", fill=True)
plt.show()
```

```python
# Strip plot 
#stripplot → quick view of data points
#Applications

'''Used when we want to visualize raw distribution of numerical data across categories.
Helpful for detecting clusters or general spread of values.'''

sns.stripplot(x="species", y="body_mass_g", data=penguins)
plt.title("Strip Plot")
plt.grid()
plt.show()
```

```python
# Swarm plot 
#A swarm plot is similar to a strip plot, but points are arranged to avoid overlap. 
#This ensures all data points are visible, making it more informative.
# swarmplot → detailed, clean distribution


sns.swarmplot(x="species", y="body_mass_g", hue="sex", data=penguins)
sns.set(style = 'whitegrid')
plt.title("Swarm plot")
plt.show()
```

```python
# Point Plot
#This method is used to show point estimates and confidence intervals using scatter plot glyphs

sns.pointplot(x="species", y="body_mass_g", hue="sex", data=penguins)
plt.show()
```

```python
# Displot
# displot → distribution (hist + kde)
sns.displot(data=penguins, x="body_mass_g", kde=True)
plt.show()
```

```python
#Rug Plot
# rugplot → individual data points
sns.rugplot(x="flipper_length_mm", data=penguins)
plt.show()
```

```python
# Clustermap
# clustermap → clustering + heatmap
penguins = sns.load_dataset("penguins")

numeric_df = penguins.select_dtypes(include=["float64", "int64"])
corr = numeric_df.corr()

sns.clustermap(corr)
plt.show()
```

```python
#JointPlot 
# jointplot → 2-variable relationship
sns.jointplot(x="flipper_length_mm", y="body_mass_g", data=penguins)
plt.show()
```

```python
#FaceGrid 
# FacetGrid → multiple subplots by category
g = sns.FacetGrid(penguins, col="species")
g.map(sns.histplot, "body_mass_g")

plt.show()
```

```python
#Regplot
#shows relationship between 2 numeric variables
#scatter points
#regression line (trend)

import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("penguins")

sns.regplot(x="flipper_length_mm", y="body_mass_g", data=penguins)
plt.show()
```

```python
#LmPlot 
#regression with grouping
# lmplot → multiple plots + grouping (figure-level)

sns.lmplot(x="flipper_length_mm", y="body_mass_g", hue="species", data=penguins)
plt.show()
```
